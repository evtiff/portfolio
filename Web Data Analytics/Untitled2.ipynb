```python
import pandas as pd
sr = pd.read_csv("Clean Data.csv")
```

```python
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer

lines_list = list(sr["review"])

sid=SentimentIntensityAnalyzer()
sslist = []
for review in lines_list:
    sslist.append(sid.polarity_scores(str(review))["compound"])
```

```python
len(sslist) == len(sr["review"])
```

```python
sr["sentiment score"] = sslist
```

```python
sr.head()
```

```python
sr[sr["app_name"] == "Dead by Daylight"].groupby(["app_name", "date"]).agg({"sentiment score": "mean"}).to_csv("DBDaggregate.csv")
```

```python
sr[sr["app_name"] == "Tom Clancy's Rainbow Six Siege"].groupby(["app_name", "date"]).agg({"sentiment score": "mean"}).to_csv("RSSaggregate.csv")
sr[sr["app_name"] == "Among Us"].groupby(["app_name", "date"]).agg({"sentiment score": "mean"}).to_csv("AUaggregate.csv")
sr[sr["app_name"] == "Grand Theft Auto V"].groupby(["app_name", "date"]).agg({"sentiment score": "mean"}).to_csv("GTAaggregate.csv")
sr[sr["app_name"] == "PLAYERUNKNOWN'S BATTLEGROUNDS"].groupby(["app_name", "date"]).agg({"sentiment score": "mean"}).to_csv("PUBGaggregate.csv")
```
